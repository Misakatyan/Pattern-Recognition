```python
import numpy as np
import matplotlib.pyplot as plt
import math
#生成样本
N=1000000
sample=np.random.normal(0,1,N)
sample.sort()
x=[i for i in range(0,N)]
```

```python
from cmath import inf

k_N=201
def estimate(k_N=101):
    x=[]
    p_x=[]
    loc=0
    # 估计
    for i in range(-5000,5000):
        x_n=i/1000
        while(sample[loc]<x_n and loc<N-1):
            loc+=1
        if(loc<(k_N-1)/2):
            x.append(x_n)
            p_x.append(loc*2/N/(sample[loc*2]-sample[0]))
        elif(loc>=N-(k_N-1)/2-1):
            x.append(x_n)
            p_x.append((N-1-loc)/N/(sample[N-1]-sample[N-1-(N-1-loc)*2]))
        else:
            x.append(x_n)
            p_x.append(k_N/N/(sample[loc+int((k_N-1)/2)]-sample[loc-int((k_N-1)/2)]))
    plt.plot(x,p_x)
    stan()
    plt.show()
```

```python
def stan():
    x_stan=[]
    y_stan=[]
    for i in range(-500,500):
        x=i/100
        y=math.exp(-x**2/2)/(math.sqrt(2*math.pi))
        x_stan.append(x)
        y_stan.append(y)
    plt.plot(x_stan,y_stan)
```

```python
for i in range(101,10001,1000):
    estimate(i)
```
