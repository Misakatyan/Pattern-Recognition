```python
import numpy as np
import matplotlib.pyplot as plt
from math import*

plt.figure(figsize=(8,3),dpi=400)

#Xy
X = np.linspace(0,7,1000)
sigma1,sigma2=0.5,1
mu1,mu2=1.5,3.5
y1=1/np.sqrt(2*pi)/sigma1*np.exp(-0.5*(X-mu1)*(X-mu1)/sigma1/sigma1)
y2=1/np.sqrt(2*pi)/sigma2*np.exp(-0.5*(X-mu2)*(X-mu2)/sigma2/sigma2)

#
plt.plot(X,y1,linewidth=3.0,linestyle='-',color='red',label="p(x|omega_1)")
plt.plot(X,y2,linewidth=3.0,linestyle='-',color='blue',label="p(x|omega_2)")
plt.legend(loc='upper left')
plt.show()
```
