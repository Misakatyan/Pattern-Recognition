```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi']=400
import numpy as np
import random
from math import*
```

每个子类的样本

```python
scale=0.5
sample_num=50
def r():
    return random.random()*0.7-0.35
x3=np.random.normal(loc=[1+r(),1+r()],scale=scale,size=[sample_num,2])
x6=np.random.normal(loc=[-1+r(),1+r()],scale=scale,size=[sample_num,2])
x1=np.random.normal(loc=[-1+r(),-1+r()],scale=scale,size=[sample_num,2])

x5=np.random.normal(loc=[1+r(),-1+r()],scale=scale,size=[sample_num,2])
x2=np.random.normal(loc=[1+r(),-3+r()],scale=scale,size=[sample_num,2])
x4=np.random.normal(loc=[-1+r(),-3+r()],scale=scale,size=[sample_num,2])
x=[]
y=[]
for i in x1:
    temp=np.array([1,i[0],i[1]])
    x.append(temp)
    y.append(np.array([1,1]))
for i in x2:
    temp=np.array([1,i[0],i[1]])
    x.append(temp)
    y.append(np.array([1,2]))
for i in x3:
    temp=np.array([1,i[0],i[1]])
    x.append(temp)
    y.append(np.array([1,3]))
for i in x4:
    temp=np.array([1,i[0],i[1]])
    x.append(temp)
    y.append(np.array([2,1]))
for i in x5:
    temp=np.array([1,i[0],i[1]])
    x.append(temp)
    y.append(np.array([2,2]))
for i in x6:
    temp=np.array([1,i[0],i[1]])
    x.append(temp)
    y.append(np.array([2,3]))
x=np.matrix(x)
y=np.matrix(y)
```

初始化每个子类的权重

```python
omega=np.random.normal(size=[13,3,1,1])
omega=np.matrix(omega)
```

判别函数

```python
def classify(x):
    outcome=[]
    for i in range(len(x)):
        y_i=np.matrix([[x[i][0]],[x[i][1]],[x[i][2]]])
        res1=-1e100
        for i in range(1,7):
            if(omega[i]*y_i>res1):
                res1=omega[i]*y_i
        res2=-1e100
        for i in range(7,13):
            if(omega[i]*y_i>res2):
                res2=omega[i]*y_i
        if(res1>res2):
            outcome.append(1)
        else:
            outcome.append(2)
    return np.array(outcome)
```

```python
x_predict=[]
y_predict=[]
x=np.array(x)
y=np.array(y)
for i in range(len(x)):
    x_predict.append(x[i][1:3])
    y_predict.append(y[i][0])
x_predict=np.array(x_predict)
y_predict=np.array(y_predict)
```

```python
def train(x,y,iteration=10):
    global omega
    for k in range(iteration):
        for i in range(len(x)):
            y_i=np.matrix([[x[i][0]],[x[i][1]],[x[i][2]]])
            res1=-1e100
            mark1,mark2=0,0
            for j in range(1,7):
                if(omega[j]*y_i>res1):
                    res1=omega[j]*y_i
                    mark1=j
            res2=-1e100
            for j in range(7,13):
                if(omega[j]*y_i>res2):
                    res2=omega[j]*y_i
                    mark2=j
            result=0
            if(res1>res2):
                result=1
            else:
                result=2
            step=0.001/(log2(k+2))
            if(result==y[i][0]):
                continue
            elif(result==1):
                omega[mark1]-=step*y_i.T
                omega[mark2]+=step*y_i.T
            else:
                omega[mark1]+=step*y_i.T
                omega[mark2]-=step*y_i.T
        if(k%200==0):
            plot_regions(x_predict,y_predict)
```

```python
def plot_regions(x, y, resolution=0.1):
    markers = ['s', 'x', 'o', '^', 'v']
    colors = ['r', 'g', 'b', 'gray', 'cyan']
    cmap = ListedColormap(colors[:len(np.unique(y))])
    '''x1_min, x1_max = x[:, 1].min() - 1, x[:, 1].max() + 1
    x2_min, x2_max = x[:, 2].min() - 1, x[:, 2].max() + 1'''
    x1_min, x1_max =-3,3
    x2_min, x2_max =-5,3
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution), np.arange(x2_min, x2_max, resolution))
    alignment =[1]*len(xx1.ravel())
    alignment=np.array(alignment)
    z = classify(np.array([alignment,xx1.ravel(), xx2.ravel()]).T)
    print(z.shape)
    z = z.reshape(xx1.shape)
    print(z.shape)
    plt.contourf(xx1, xx2, z, alpha=0.4, cmap=cmap)
 
    for idx, cc in enumerate(np.unique(y)):
        plt.scatter(x=x[y == cc, 0],
                    y=x[y == cc, 1],
                    alpha=0.6,
                    c=cmap(idx),
                    edgecolor='black',
                    marker=markers[idx],
                    label=cc)
    plt.show()
plot_regions(x_predict,y_predict)
train(x,y,iteration=50000)
```
